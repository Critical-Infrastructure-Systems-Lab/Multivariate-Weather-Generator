```python
import pandas as pd
import numpy as np
import xarray as xr
import os
```

```python
# netCDF to csv
for i in ["merged_0.25_Precip.nc",
          "merged_0.25_Tmax.nc",
          "merged_0.25_Tmin.nc",
          "merged_0.25_Wind.nc"]:
    netcdf_file_name = i
    netcdf_file_in = netcdf_file_name
    csv_file_out = netcdf_file_name[:-3] + '.csv'
    ds = xr.open_dataset(netcdf_file_in)
    df = ds.to_dataframe()
    df.to_csv(csv_file_out)
```

```python
# An example of how to process each dataset
df = pd.read_csv("merged_0.25_Wind.csv")
df = df[df["bnds"] == 0]
df = df.fillna(value = 0.0)
df = df[["time", "lon", "lat", "wind"]]
df.columns = ["Date", "Longitude", "Latitude", "wind"]
```

```python
# Original training dataset without wind
training = pd.read_csv("Training_no_wind.csv")
```

```python
# An example of column binding different variables
training = pd.concat([training, (df["wind"].reset_index())], axis=1)
training = training[["Date", "Longitude", "Latitude", "precipitation", "t_min", "t_max", "wind"]]
```
